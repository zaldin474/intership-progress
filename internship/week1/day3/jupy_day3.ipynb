Array indexing and slicing:

```python
a = np.linspace(1,16,16)
a = np.reshape(a,(4,4))

print(a.size)
print(a.shape)
print(a.ndim)

print(a[:,2]) # from all raws, column 2
print("----------------------")

print(a[0,:]) # from row 0, all columns
print("----------------------")

print(a[0:2,0:2]) # from row 0 to 1, column 0 to 1
print("----------------------")

print(a[3,:]) # from row 3, all columns
print("----------------------")

print(a[0:2,1:3]) # from row 0 to 1, column 1 to 2
print("----------------------")

print (a[a>5]) # all elements greater than 5 (boolean masking)
```

Vectorized operations:

```python
a= np.arange(1,17,1)
a= np.reshape(a,(4,4))

b=np.array([1,2,3,4])

print(a+b)                      #broadcasting
print("----------------------")
print(a*b)
print("----------------------")
print(a@b) # matrix multiplication
print("----------------------")
print(a*2) # multiply each element by 2
print("----------------------")
print(a.mean()) # mean of all elements
print("----------------------")
print(a.mean(axis=0)) # mean of each column
print("----------------------")
print(a.mean(axis=1)) # mean of each row
print("----------------------")
print(a.sum()) # sum of all elements
print("----------------------")
print("----------------------")
```
