```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

simple line plot

```python
x = np.linspace(0, 10, 5)
y = np.array([1, 3, 6, 9, 12])
plt.plot(x,y)
plt.plot(y,x)
plt.xlabel('time')
plt.ylabel('value')
plt.title('Line plot')
plt.show()
```

scatter vs bar vs hist graph

```python
x = np.linspace(0, 10, 5)
y = np.array([1, 3, 6, 9, 12])
z = np.sin(x)
plt.scatter(x,y)
plt.bar(y,x)
plt.hist(x,y)
plt.xlabel('time')
plt.ylabel('value')
plt.title('Line plot')
plt.show()
```

subplots:

```python
fig , ax = plt.subplots(figsize=(10,4))
ax.plot(x,y)
ax.set_xlabel('velocity')
ax.set_ylabel('time')
ax.set_title('line plot')
plt.show()
```

```python
# Creates a figure with 1 row and 2 columns of graphs
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

ax1.plot(x, y)
ax1.set_title('Velocity')

ax2.plot(x, z) # Assuming 'z' is another list of data
ax2.set_title('Temperature')
plt.grid(True)
plt.show()
```
