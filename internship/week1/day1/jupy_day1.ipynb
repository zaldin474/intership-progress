same but takes input from user

```python
# Interactive compound interest calculator
initial_amount = float(input("Enter the initial amount: "))
interest = float(input("Enter the interest rate (as decimal, e.g., 0.05 for 5%): "))
payment_period = int(input("Enter the number of payment periods: "))

compound_amount = initial_amount * ((1 + interest) ** payment_period)
print("initial_amopunt: ", initial_amount)
print(f"Interest Rate: {interest * 100}%")
print("Payment Periods: ",payment_period)
print("Compound Amount: ", compound_amount)
```
