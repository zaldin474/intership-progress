Task 4:

This list the mean of tackles, interceptions, clearances and blocks grouped by position.

```python
df.groupby("position").agg({"tackles": "mean" , "interceptions": "mean", "clearances": "mean", "blocks": "mean"}).sort_values(by="tackles", ascending=False)
```

This one list expected-goals, assistes and shots on target grouped by teams. (it might not be real life accurate but it is accurate to the dataSet)

```python
df.groupby("team").agg({"expected_goals_xg": "sum", "assists": "sum", "shots_on_target": "sum"}).sort_values(by="expected_goals_xg", ascending=False)
```

this one shows the number of goals scored based on the preferred_foot. 
it reflects a real-life observation as the number of players with right foot as the prefered foot is larger, the goal corelated are more.

```python
df.groupby("preferred_foot").agg({"goals": "sum", "assists": "sum", "shots_on_target": "sum"}).sort_values(by="goals", ascending=False)
```
