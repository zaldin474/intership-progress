Selecting and filtering:

```python
df[["Pulse", "Calories"]]  # multiple columns (DataFrame)
```

```python
df[df["Duration"] < 60] #filter rows based on a condition (Duration < 60) 
```

Cleaning Data:

```python
df.isnull().sum()  # check for missing values (NaN) in the DataFrame per column
```

```python
df["Calories"].fillna(df["Calories"].mean(), inplace=True) # fill missing values in Calories column with the mean of that column
```

```python
df.dropna(inplace=True) # drop rows with any missing values
```

```python
df.drop_duplicates(inplace=True) # drop duplicate rows
```

Grouping & Aggregation:

```python
df.groupby("Duration")["Pulse"].mean() # group by Duration and calculate the mean of Pulse
```

```python
df.groupby("Duration")[["Pulse", "Maxpulse", "Calories"]].mean()
```

```python
df.groupby("Duration").agg({"Pulse": "mean", "Maxpulse": "max", "Calories": "mean"}) # group by Duration and aggregate multiple columns with different functions
```
