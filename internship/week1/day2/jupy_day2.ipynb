  
    Calculate the mean, minimum, and maximum of a list of numbers, and produce an output as a dict data type.

```python
def mean_min_max():
    numbers = [float(x) for x in input("Enter numbers separated by spaces: ").split()]
    d = {}
    d['mean']= sum(numbers)/len(numbers)
    d['min'] = min(numbers)
    d['max'] = max(numbers)
    return d   

mean_min_max()
```
